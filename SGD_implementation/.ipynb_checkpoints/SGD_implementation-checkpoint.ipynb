### Stochastic Gradient Descent

#### objective : Implementing SGD and get result near to Sklearn SGD implementation

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.datasets import load_boston
from random import seed
from random import randrange
from csv import reader
from math import sqrt
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from sklearn.linear_model import SGDRegressor
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X = load_boston().data
Y = load_boston().target
```

```python
#divide data into train test split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3 )
```

```python
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#train SGD model
clf = SGDRegressor()
clf.fit(x_train, y_train)
print(mean_squared_error(y_test, clf.predict(x_test)))
```

```python
import pickle
with open('sklearn_sgd_model.pkl','wb') as file:
    pickle.dump(clf,file)
```

```python
import random 
```

```python
#Calculate MSE
def get_mse(w,b,data,label):
   sqr_loss =  [(label[idx] - (np.dot(w.T,x) + b))**2 for idx,x in enumerate(data)]
   return sum(sqr_loss)/len(label)
```

```python
learning_rate = 0.001
#initialize weight and bias of model
mu,sigma = 0,0.1
w = np.random.normal(mu,sigma,13)
b = random.random()

iteration_loss_train = []
w_new,b_new = np.zeros((13,)),0.0

while True:
    train_loss = 0.0
    w_loss,b_loss=0.0,0.0
    # Randomly generate number to select random points
    random_indexes = np.random.randint(2,x_train.shape[0],1)[0]
    rand_xtrain = x_train[:random_indexes]
    rand_ytrain = y_train[:random_indexes]
    for idx,x in enumerate(rand_xtrain):
        w_loss += ((x * (-2))*(y_train[idx] - (np.dot(w.T,x) + b))) 
        b_loss += (-2)*(y_train[idx] - (np.dot(w.T,x) + b))
    w_new = w - learning_rate * w_loss
    b_new = b - learning_rate * b_loss
    train_loss = get_mse(w_new,b_new,rand_xtrain,rand_ytrain)
    #stop when two consecutive weights are same
    if (w==w_new).all():
        iteration_loss_train.append(train_loss)
        w = w_new
        b = b_new
        break
    iteration_loss_train.append(train_loss)
    w = w_new
    learning_rate = learning_rate/2
    b = b_new
```

```python
#Calculate test MSE
get_mse(w,b,x_test,y_test)
```

```python
import pickle
classifier = (w,b)
with open('custom_SVD_clf.pkl','wb') as file:
    pickle.dump(classifier,file)
```

```python
# Sklearn and Implemented SGD weight comparison
print("Learning Rate : ",learning_rate)
plt.figure(figsize=(12, 9))
plt.plot(range(len(w)), clf.coef_, color='green', lw=2, label='Sklearn SGD W')
plt.plot(range(len(w)), w, color='red', lw=2, label='Implemented SGD W')
plt.xlabel('Features')
plt.ylabel('Weights')
plt.legend(loc="upper right")
plt.title("Sklearn and Implemented SGD weight comparison")
plt.grid(True,color='black')
plt.show()
```

```python
# Prices vs Predicted Values plot
#Sklearn SGD
plt.figure(figsize=(12, 9))
plt.scatter(y_test, clf.predict(x_test))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Prices vs Predicted Values: Sklearn SGD")
plt.show()

#Implemented SGD
plt.figure(figsize=(12, 9))
plt.scatter([y_test], [(np.dot(x_test, w) + b)])
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Prices vs Predicted Values: Implemented SGD")
plt.show()
```
